## A Note on Original Thinking

The core ideas in MicroWorld emerged from a simple question:

> *If I had to explain what financial markets ARE — not just how to model them — what would I say?*

The answer: markets are a **multi-level, multi-timescale, multi-objective game** between agents who simultaneously model each other, discover and destroy each other's alpha, and occasionally cooperate through institutional structures.

Every piece of mathematics in this codebase follows from taking that description seriously.
