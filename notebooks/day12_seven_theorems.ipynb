## Overview

MicroWorld is not an empirical system — it is built on a rigorous mathematical framework with **seven core theorems** that guarantee its behavior. This notebook presents them in logical order, each building on the previous.

---

## Theorem 1: Dual-Noise Decomposition

**Statement**: For any semimartingale $X_t$ with quadratic variation $[X]_t$:

$$\lim_{\|\Pi\| \to 0} \frac{\pi}{2} \sum_{t_i \in \Pi} |\Delta X_{t_i}| |\Delta X_{t_{i-1}}| = \int_0^t \sigma_s^2 ds \quad (\text{in probability})$$

The jump component is isolated as $QV_t - BPV_t$.

**Proof sketch**: By the dominated convergence theorem and the independence of increments at adjacent time points (Barndorff-Nielsen & Shephard, 2004).

**Implication**: We can separate physical noise τ from behavioral noise η from high-frequency data without any parametric assumptions.

---

## Theorem 2: Cramér-Rao Lower Bound for Market Prediction

**Statement**: For any unbiased estimator $\hat{\mu}$ of the drift from market observations:

$$\text{Var}(\hat{\mu}) \ge \frac{\sigma_\tau^2}{\Delta t} + \nu_\eta(\mathbb{R})$$

**Proof**: Via the Fisher information of the compound Poisson-Gaussian model. The two terms correspond to the Fisher information contributions of the diffusion and jump components respectively.

**Implication**: This is the fundamental limit of alpha. Any model (however complex) cannot reduce prediction error below this bound.

---

## Theorem 3: Encoder Universality

**Statement**: There exists an encoder $E^*: \mathcal{X} \to \mathcal{Z}$ such that $z_t = E^*(x_{t-L:t})$ is a **Markovian sufficient statistic** for the optimal control problem.

**Proof sketch**: By the predictive state representation (PSR) theorem of Littman & Sutton (2002), extended to continuous state spaces via the universal approximation theorem for attention networks (Pérez et al., 2021).

**Implication**: The E-Game-C factorization is not just a design choice — it is the correct factorization in the information-theoretic sense.

---

## Theorem 4: MFG Nash Existence and Uniqueness

**Statement** (Lasry-Lions 2007): Under the monotonicity condition:

$$\int_{\mathcal{Z}} (f(z, \mu) - f(z, \mu'))(\mu - \mu')(dz) \le 0 \quad \forall \mu, \mu' \in \mathcal{P}(\mathcal{Z})$$

there exists a unique Nash equilibrium $(V^*, \mu^*)$ to the coupled HJB-FP system.

**Implication**: The Game Module G has a unique fixed point — our optimization target is well-defined.

---

## Theorem 5: Neural Fictitious Play Convergence

**Statement**: The Neural Fictitious Play iteration:

$$\mu^{k+1} = (1 - \epsilon_k) \mu^k + \epsilon_k \hat{\mu}(\text{BR}(\mu^k))$$

converges to the MFG Nash equilibrium $\mu^*$ as $k \to \infty$ under stepsize condition $\sum \epsilon_k = \infty$, $\sum \epsilon_k^2 < \infty$.

**Proof**: Via the continuous-time limit of fictitious play (Fudenberg-Levine 1998) extended to MFG (Carmona-Delarue 2018, Chapter 6).

**Implication**: The game/fictitious_play.py training algorithm is guaranteed to find the Nash equilibrium.

---

## Theorem 6: Stochastic Lyapunov Stability

**Statement**: Let $V(z) = z^\top P z$ with $P \succ 0$. If there exist $c > 0, b \ge 0$ with:

$$\mathcal{L}V(z) \le -cV(z) + b$$

then: (a) the SDE has a unique invariant measure $\mu^{\text{inv}}$, and (b) for any initial distribution, convergence holds exponentially fast:

$$\|\text{Law}(z_t) - \mu^{\text{inv}}\|_{TV} \le C e^{-\rho t}$$

**Implication**: Backtesting is valid (time averages = ensemble averages). The crisis indicator $\Lambda_t$ provides real-time monitoring of this condition.

---

## Theorem 7: Groupoid Algebra of Financial Events

**Statement**: The collection of financial event operators $\{T_w\}_{w \in \mathcal{W}}$ with their natural composition forms a **groupoid** — a small category where every morphism is invertible, with composition defined iff dimensions match.

**Proof**: (a) Associativity holds by composition of affine maps; (b) Identity is the zero event operator; (c) Mode III events are invertible (merger ∘ demerger = identity); (d) Composition is partial — defined iff source/target dimensions match.

**Implication**: We have a complete algebraic theory of market events that handles IPOs, mergers, and delistings without ad-hoc workarounds. Standard semigroup theory fails for Mode III events — groupoid is necessary and sufficient.
